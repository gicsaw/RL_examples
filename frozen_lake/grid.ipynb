```python
import gym
import numpy as np
from gym.envs.toy_text.frozen_lake import FrozenLakeEnv
```

```python
env =FrozenLakeEnv(is_slippery=False,map_name="8x8")
#env =FrozenLakeEnv(is_slippery=False,map_name="4x4")
env.render()
```

```python
max_episodes=200
max_step=100
lr=0.5
gamma=0.99
epsilon0 = 0.1

# initialize
Q = np.zeros([env.observation_space.n,env.action_space.n])
epsilon=epsilon0
min_step=max_step
min_step_episode=max_episodes
for episode in range(max_episodes):
    epsilon=epsilon*0.99
    state = env.reset()
    total_reward = 0
    state_list = [state]
    win=0
    for step in range(0,max_step):
        if np.random.rand()>epsilon :
            action=np.argmax(Q[state,:])
        else:
            action=np.random.choice(env.action_space.n)
        
        state_next,reward,game_over,_ = env.step(action)
    
        if reward==0 : 
            reward=-0.01
        if reward!=1.0 and game_over==1:
            reward=-1
#            game_over=0
        # Q-Learning
        Q[state,action]= Q[state,action]+lr*(reward+gamma* np.max(Q[state_next,:])-Q[state,action])
        state=state_next
        total_reward=total_reward+reward
        state_list.append(state)
        if game_over and reward==1:
            win=1
            if step< min_step:
                min_step=step
                min_step_path=state_list
                min_step_episode=episode
        if game_over==1:
            break
            

    line_out="Episode: %d, step: %d, total_reward: %f" %(episode, step, total_reward)
    print(line_out)
    if reward==1 :
        print(state_list)
print( "min step: ", min_step, "min step episode", min_step_episode)
print("shortest path: ", min_step_path)
#print(Q)
```
